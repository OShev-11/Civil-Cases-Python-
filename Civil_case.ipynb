```python
def civil_cases():
    cases = ['2-1/26', '2-2/26', '5-1/26', '5-2/26']
    case_type = ['Гражданское','Гражданское',  'Административное', 'Административное']
    status = ['Выиграно', 'Проиграно', 'Проиграно', 'Выиграно']
    case_list = []

    for i in range(len(cases)):
        if status[i] == 'Проиграно' and case_type[i] == 'Гражданское':
           case_list.append(cases[i])

    return case_list

print(civil_cases())
```
